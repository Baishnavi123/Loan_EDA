# Part-4

- Steps in outlier analysis
- step1: Find Q1,Q2,Q3
    - np.percentile(column data,q)
- step2: calculate the IQR
    - IQR=Q3-Q1
- step3: calculate lower boundary and upper boundary
    - lb: Q1-1.5IQR
    - ub: Q3+1.5IQR
- step4: Find the outliers df
    - c1: column data < lb
    - c2: column data > ub
    - c: apply the main condition
    - main data[c]

# Impute with median values

```python
loan_df['LoanAmount'].median()
```

```python
Q1=np.percentile(loan_df['LoanAmount'],25)
Q3=np.percentile(loan_df['LoanAmount'],75)
IQR=Q3-Q1
lb=Q1-1.5*IQR
ub=Q3+1.5*IQR
median_value=[]
median=loan_df['LoanAmount'].median()
for i in loan_df['LoanAmount']:
    if i <lb or i >ub:
        median_value.append(median)
    else:
        median_value.append(i)
loan_df['LoanAmount_new']=median_value
loan_df['LoanAmount_new']

# Here in the Dataset wherever there is null value it will be filled with the median value of the dataset
# to get rid of outliers
```

- above replace one we use a traditional approach
- for loop, list,if-else
- the same we can get by using np.where method
- np.where(con,True,False)
- will take three arguement
- condition
    - con=d['Num']>2
- True value
    - t=100
- False value
    - f=d['Num']

# ##############################################################

```python
Q1=np.percentile(loan_df['LoanAmount'],25)
Q3=np.percentile(loan_df['LoanAmount'],75)
IQR=Q3-Q1
lb=Q1-1.5*IQR
ub=Q3+1.5*IQR
median=loan_df['LoanAmount'].median()
c1=loan_df['LoanAmount']>ub
c2=loan_df['LoanAmount']<lb
con=c1|c2

l=np.where(con,median,loan_df['LoanAmount'])
loan_df['LoanAmount']=l

plt.boxplot(loan_df['LoanAmount'])
plt.show()
```
